## creating new table for software aka photo editor

```python
-- checking data

SELECT software
FROM 
(SELECT software,
		count(*) 
FROM photo_exif_top100
WHERE lower(software)  LIKE '%photoshop%' AND lower(software) NOT LIKE '%lightroom%'
GROUP BY software) AS blub;

SELECT *
FROM photo_exif_top100
WHERE lower(software)  LIKE '%windows%';

SELECT *
FROM photo_exif_top100
WHERE lower(software)  LIKE '%lightroom%';

SELECT DISTINCT(software), count(*)
FROM photo_exif_top100
GROUP by software 
ORDER BY  count DESC  ;

SELECT count(*)
FROM photo_exif_top100 pet 
WHERE software IS NOT NULL ;

--------------------------
-- just in case

DROP TABLE software_grouped ;
-----------------

SELECT sum(total) AS photoshop
FROM (SELECT software,
		count(*) AS total
FROM photo_exif_top100
WHERE lower(software)  LIKE '%photoshop%' AND lower(software) NOT LIKE '%lightroom%'
GROUP BY software) AS blub;

SELECT sum(total) AS lightroom
FROM (SELECT software,
		count(*) AS total
FROM photo_exif_top100
WHERE lower(software)  LIKE '%lightroom%'
GROUP BY software) AS blub;

SELECT sum(total) AS windows_pe
FROM (SELECT software,
		count(*) AS total
FROM photo_exif_top100
WHERE lower(software)  LIKE '%windows photo editor%'
GROUP BY software) AS blub;
--


-- create new table for photo editing software

CREATE TABLE software_grouped
  AS (SELECT id, software, make, model FROM photo_exif_top100);

 -- group photo editing programs
 
 ALTER TABLE software_grouped 
 ADD photo_editor text;

UPDATE software_grouped 
SET photo_editor = 'Photoshop'
WHERE lower(software)  LIKE '%photoshop%' AND lower(software) NOT LIKE '%lightroom%' ;

UPDATE software_grouped 
SET photo_editor = 'Lightroom'
WHERE lower(software)  LIKE '%lightroom%';

SELECT distinct(photo_editor),count(*) 
FROM software_grouped sg 
GROUP BY photo_editor ;

UPDATE software_grouped 
SET photo_editor = 'Windows Photo Editor'
WHERE lower(software)  LIKE '%windows photo editor%';

UPDATE software_grouped 
SET photo_editor = 'Ver.1.x'
WHERE lower(software)  LIKE 'ver.1.%' or lower(software)  LIKE 'ver.1.00';

UPDATE software_grouped 
SET photo_editor = '16.x'
WHERE lower(software)  LIKE '16.%';


UPDATE software_grouped 
SET photo_editor = 'PhotoScape'
WHERE lower(software)  LIKE 'photoscape';

UPDATE software_grouped 
SET photo_editor = 'Gimp'
WHERE lower(software)  LIKE '%gimp%';

UPDATE software_grouped 
SET photo_editor = 'Capture One'
WHERE lower(software)  LIKE '%capture one%';

UPDATE software_grouped 
SET photo_editor = 'Digital Photo Professional'
WHERE lower(software)  LIKE '%digital photo professional%';

UPDATE software_grouped 
SET photo_editor = 'Sony ILCE'
WHERE lower(software)  LIKE '%ilce-%';

UPDATE software_grouped 
SET photo_editor = 'Picasa'
WHERE lower(software)  LIKE '%picasa%';

UPDATE software_grouped 
SET photo_editor = 'Microsoft Windows Photo Viewer'
WHERE lower(software)  LIKE '%microsoft windows photo viewer%'

UPDATE software_grouped 
SET photo_editor = 'Snapseed'
WHERE lower(software)  LIKE '%snapseed%'


SELECT distinct(photo_editor),count(photo_editor), make
FROM software_grouped sg 
--WHERE software IS NOT NULL 
GROUP BY photo_editor, make ;

-- add editor version again

ALTER TABLE software_grouped  
ADD editor_version TEXT;

UPDATE software_grouped 
SET editor_version = software
WHERE photo_editor IS NOT NULL ;

-- add column if software is jused or not

ALTER TABLE software_grouped 
ADD software_used TEXT;

UPDATE software_grouped 
SET software_used = 'True'
WHERE software IS NOT NULL;

UPDATE software_grouped 
SET software_used = 'False'
WHERE software IS NULL;

-- add operating system

ALTER TABLE software_grouped 
ADD os TEXT;

UPDATE software_grouped 
SET os = 'windows'
WHERE lower(software) LIKE '%windows%';

UPDATE software_grouped 
SET os = 'mac'
WHERE lower(software) LIKE '%macintosh%';

SELECT DISTINCT os, count(*)
FROM software_grouped sg 
GROUP BY os;

```
