In this notebook we are using the image IDs we gathered in our exif csv file, select an image size and download images to a folder.

```python
# Python modules
import flickrapi
import pandas as pd
import os
import json
from dotenv import load_dotenv
load_dotenv()
import numpy as np
import subprocess

# Import additional functions
from flickr_functions import df_remove_dupes

# Keys needed for API access
api_key = os.getenv('flickr_api_key')
api_secret = os.getenv('flickr_api_secret')

api_key = '8eda6fcb3fd5a4de7b5714cd109221ca'
api_secret = '8d63ce31d49cdfae'

# Flickr API object
flickr = flickrapi.FlickrAPI(api_key, api_secret, format='parsed-json')

# Data directory used to store CVS files
data_dir = './data/'

# Load EXIF dataframe from CVS file
df = pd.read_csv(data_dir + 'df_photo_exif_final.csv', index_col=[0])

# Run command function
def runcmd(cmd, verbose = False, *args, **kwargs):

    process = subprocess.Popen(
        cmd,
        stdout = subprocess.PIPE,
        stderr = subprocess.PIPE,
        text = True,
        shell = True
    )
    std_out, std_err = process.communicate()
    if verbose:
        print(std_out.strip(), std_err)
    pass
```

```python
# Select a range of image IDs from our exif file
ids = df['id'].sample(100).astype(int).to_list()
```

```python
# Get a list of available sizes
flickr.photos.getSizes(photo_id = ids[0])
```

### Iterate through IDs and download image

```python
for id in ids:
    # Get list of available image sizes
    try:
        sizes = flickr.photos.getSizes(photo_id = id)

        # Iterate through sizes
        for size in sizes.get('sizes').get('size'):

            # Search for a certain image size
            if size.get('label') == 'Small 400':
                print('Downloading: ' + size.get('source'))

                # Download the image using runcmd function
                runcmd('wget ' + size.get('source') + ' -P ./images', verbose = True)
    except flickrapi.exceptions.FlickrError as ex:
        # If accessing the API fails, print something out
        print(f'!!! Error code: {ex.code} for id {id}')
```
