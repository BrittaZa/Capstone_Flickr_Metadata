### Get all camera makes and models known to flickr

In this notebook we are using the flickr API to get data about all of the camera manufacturers and their models in the flickr database.

```python
# Python modules
import flickrapi
import pandas as pd
import os
import json
from dotenv import load_dotenv
load_dotenv()

# Import additional functions
from flickr_functions import df_remove_dupes

# Keys needed for API access
api_key = os.getenv('flickr_api_key')
api_secret = os.getenv('flickr_api_secret')

#api_key = 'f47b5d3f72fd951303247c536c9168bd'
#api_secret = 'fa4e2668b6c914d1'

# Flickr API object
flickr = flickrapi.FlickrAPI(api_key, api_secret, format='parsed-json')

# Data directory used to store CVS files
data_dir = './data/'
```

```python
# Get the list of brand IDs and names
try:
    brands = flickr.cameras.getBrands()
except flickrapi.exceptions.FlickrError as ex:
    print("Error code: %s" % ex.code)

df_flickr_brands = pd.DataFrame(columns = ['id', 'name'])

# Iterate through the brands and get all of the models
for brand in brands['brands']['brand']:
    
    try:
        models = flickr.cameras.getBrandModels(brand = brand.get('id'))

        # Iterate through the models of one manufacturer
        for model in models.get('cameras').get('camera'):

            # Create a temporary dict
            dict_tmp = {'brand_id': brand.get('id'),
                        'brand': brand.get('name'),
                        'model_id': model.get('id'),
                        'name': model.get('name').get('_content')
                        }
            
            # Sometimes image links could be empty, so we need to handle that
            if model.get('images'):
                photo_small = {'image_small': model.get('images').get('small').get('_content')}
                photo_large = {'image_large': model.get('images').get('large').get('_content')}
            else:
                photo_small = {'image_small': ''}
                photo_large = {'image_large': ''}
                
            dict_tmp.update(photo_small)
            dict_tmp.update(photo_large)

            # Create row in dataframe
            df_flickr_models.loc[len(df_flickr_models)] = dict_tmp

    except flickrapi.exceptions.FlickrError as ex:
        print("Error code: %s" % ex.code)
```

```python
df_flickr_models.to_csv('./data/df_flickr_models.csv')
```
